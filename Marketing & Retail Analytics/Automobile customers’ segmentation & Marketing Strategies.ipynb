### Customer Segmentation using RFM analysis

```python
rfm_df = df[['CUSTOMERNAME','SALES','DAYS_SINCE_LASTORDER']]
```

```python
recency = rfm_df.groupby('CUSTOMERNAME').min()['DAYS_SINCE_LASTORDER'].reset_index()
freq = rfm_df['CUSTOMERNAME'].value_counts().reset_index().rename(columns={'CUSTOMERNAME':'frequency','index':'CUSTOMERNAME'})
monetary = rfm_df.groupby('CUSTOMERNAME').sum()['SALES'].reset_index()
```

```python
rfm1 = pd.merge(left = recency, right = freq, left_on = 'CUSTOMERNAME', right_on = 'CUSTOMERNAME', how = 'left')
rfm2 = pd.merge(left = rfm1, right = monetary, left_on = 'CUSTOMERNAME', right_on = 'CUSTOMERNAME', how = 'left')

wdf = rfm2.rename(columns={'DAYS_SINCE_LASTORDER':'recency','SALES':'monetary'})
```

```python
rfm_df = wdf.drop(['CUSTOMERNAME'], axis = 1)
rfm_df.head()
```

```python
from crm_rfm_modeling import rfm
from crm_rfm_modeling.rfm import RFM
```

```python
model = RFM(weights=(0.33, 0.33, 0.34))
```

```python
model.fit(rfm_df)
```

```python
model.summary_statistics()
```

```python
pd.DataFrame(model.cutoffs)
```

```python
df_fitted=model.fitted_data
df_fitted.head()
```

```python
df_fitted.sort_values(['recency_scores','frequency_scores','monetary_scores'],ascending=[False,False,False]).head(10)
```

#### K-Means Clustering

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 
kdf = sc.fit_transform(wdf[['recency','frequency','monetary']]) 
```

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(n_clusters = 4,random_state=1)
k_means.fit(kdf)
labels = k_means.labels_
```

```python
wdf['cluster'] = labels
```

```python
wdf.head(5)
```

```python
final_df = pd.concat([wdf, df_fitted[['recency_scores','frequency_scores','monetary_scores']]],axis = 1)
```

```python
final_df.to_csv("FINAL_CLUSTERS.csv")
```

```python
final_df.groupby('cluster').mean()
```

```python
final_df[(final_df['recency_scores'] == 5) & (final_df['frequency_scores'] == 5) & (final_df['monetary_scores'] == 5)]
```
