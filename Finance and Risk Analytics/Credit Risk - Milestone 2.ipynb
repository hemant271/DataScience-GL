### 1.1 Outlier Treatment

```python
plt.figure(figsize = (25,25));
df.boxplot();
plt.xticks(rotation=90);
```

```python
# Outlier Proportion
def outlier_values(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range 
```

```python
num_cols = df.select_dtypes(include = ['int64','float64']).columns; # Gives all columns
for i in num_cols:
    ll,ul = outlier_values(df[i])
    oc = df[(df[i]<ll) | (df[i]>ul)][i].count()
    total = df[i].count()
    print("{0}\t{1}\t{2}".format(i,oc,oc*100/total))
```

```python
#Fixing Outliers
for i in num_cols:
    ll,ul = outlier_values(df[i])
    oc = df[(df[i]<ll) | (df[i]>ul)][i].count()
    total = df[i].count()
    if((oc/total) >= 0.15):
        df.drop(i, axis = 1, inplace = True)
    else:
        df[i] = np.where(df[i] < ll, ll, df[i])
        df[i] = np.where(df[i] > ul, ul, df[i])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

```python
for i in range(0,20):
    X = df.drop(['Default','Co_Name'], axis = 1)
    vif = calc_vif(X).sort_values(by = 'VIF', ascending = False).reset_index().drop('index',axis=1)
    if(vif[vif['VIF'] > 5].count()[0] > 0):
        print(vif.variables[0], vif.VIF[0])
        df.drop(vif.variables[0], axis = 1, inplace = True)
    if(vif[vif['VIF'] > 5].count()[0] == 0):
        break
```
